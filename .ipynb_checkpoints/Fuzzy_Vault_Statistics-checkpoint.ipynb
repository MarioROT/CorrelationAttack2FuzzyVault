### Resultados Sin Translaciones

```python
rsl10md = pd.read_csv('TabResCorr10mindistSoloGiro.csv')
rsl10md = rsl10md.T.reset_index()
rsl10md.columns= ['Distancia', 'Puntos encontrados promedio', 'Puntos genuinos encontrados promedio', 'Número de emparejamientos que se podrían vulnerar', 'Iteraciones promedio necesarias para vulnerar']
rsl10md = rsl10md.round({'Distancia':3, 'Puntos encontrados promedio':3, 'Puntos genuinos encontrados promedio':3, 'Número de emparejamientos que se podrían vulnerar':3, 'Iteraciones promedio necesarias para vulnerar':3})

rsl20md = pd.read_csv('TabResCorr10_3000_6000mindistSoloGiro.csv')
rsl20md = rsl20md.T.reset_index()
rsl20md.columns= ['Distancia', 'Puntos encontrados promedio', 'Puntos genuinos encontrados promedio', 'Número de emparejamientos que se podrían vulnerar', 'Iteraciones promedio necesarias para vulnerar']
rsl20md = rsl20md[1:]
rsl20md = rsl20md.round({'Distancia':3, 'Puntos encontrados promedio':3, 'Puntos genuinos encontrados promedio':3, 'Número de emparejamientos que se podrían vulnerar':3, 'Iteraciones promedio necesarias para vulnerar':3})

rsl30md = pd.read_csv('TabResCorr10_6000_9000mindistSoloGiro.csv')
rsl30md = rsl30md.T.reset_index()
rsl30md.columns= ['Distancia', 'Puntos encontrados promedio', 'Puntos genuinos encontrados promedio', 'Número de emparejamientos que se podrían vulnerar', 'Iteraciones promedio necesarias para vulnerar']
rsl30md = rsl30md[1:]
rsl30md = rsl30md.round({'Distancia':3, 'Puntos encontrados promedio':3, 'Puntos genuinos encontrados promedio':3, 'Número de emparejamientos que se podrían vulnerar':3, 'Iteraciones promedio necesarias para vulnerar':3})

frames = [rsl10md, rsl20md, rsl30md]

dfs = pd.concat(frames)
dfs = dfs.reset_index().drop(['index'], axis=1)
dfs
```

### Resultados Con Translaciones

```python
rt10md = pd.read_csv('TabResCorr10mindist250trans.csv')
rt10md = rt10md.T.reset_index()
rt10md.columns= ['Distancia', 'Puntos encontrados promedio', 'Puntos genuinos encontrados promedio', 'Número de emparejamientos que se podrían vulnerar', 'Iteraciones promedio necesarias para vulnerar']
rt10md = rt10md.round({'Distancia':3, 'Puntos encontrados promedio':3, 'Puntos genuinos encontrados promedio':3, 'Número de emparejamientos que se podrían vulnerar':3, 'Iteraciones promedio necesarias para vulnerar':3})

rt20md = pd.read_csv('TabResCorr5mindist1000trans.csv')
rt20md = rt20md.T.reset_index()
rt20md.columns= ['Distancia', 'Puntos encontrados promedio', 'Puntos genuinos encontrados promedio', 'Número de emparejamientos que se podrían vulnerar', 'Iteraciones promedio necesarias para vulnerar']
# rt20md = rt20md[1:]
rt20md = rt20md.round({'Distancia':3, 'Puntos encontrados promedio':3, 'Puntos genuinos encontrados promedio':3, 'Número de emparejamientos que se podrían vulnerar':3, 'Iteraciones promedio necesarias para vulnerar':3})

rt30md = pd.read_csv('TabResCorr3mindist1500trans.csv')
rt30md = rt30md.T.reset_index()
rt30md.columns= ['Distancia', 'Puntos encontrados promedio', 'Puntos genuinos encontrados promedio', 'Número de emparejamientos que se podrían vulnerar', 'Iteraciones promedio necesarias para vulnerar']
# rt30md = rt30md[1:]
rt30md = rt30md.round({'Distancia':3, 'Puntos encontrados promedio':3, 'Puntos genuinos encontrados promedio':3, 'Número de emparejamientos que se podrían vulnerar':3, 'Iteraciones promedio necesarias para vulnerar':3})

frames = [rt10md, rt20md, rt30md]

dfs = pd.concat(frames)
dfs = dfs.reset_index().drop(['index'], axis=1)
dfs
```

```python
rt10md['Puntos encontrados promedio']
```

```python
labels = ['Distance: {}'.format(round(float(i))) for i in dfs['Distancia']]
data =  list(dfs['Puntos encontrados promedio'])
title = 'Genuine + Chaff Matching Points'
donutPlot(data,labels,title)

cp = ['#8C6D31', '#ffdd6b', '#e9e2c9', '#dcae52', '#af7132', '#8C9363', '#637939', '#AD494A', '#E7969C', '#C4CBB9']
colors = []
for i in range(len(dfs)):
    if i < len(rt10md):
        colors.append(cp[5])
    elif i < len(rt10md)+len(rt20md):
        colors.append(cp[3])
    else:
        colors.append(cp[2])
xticks = [round(float(i)) for i in dfs['Distancia']]
data = {'MatchingPoints':list(dfs['Puntos encontrados promedio'])}
title = 'Genuine + Chaff Matching Points'
groupedBarPlot(data,xticks,title, legend = True,figsize=(25,10),width = 0.7,axislabels=['Distances','Matching Points'], colors = colors)
```

```python
colors = []
for i in range(len(dfs)):
    if i < len(rt10md):
        colors.append('blue')
    elif i < len(rt10md)+len(rt20md):
        colors.append('green')
    else:
        colors.append('red')
        
colors
```
