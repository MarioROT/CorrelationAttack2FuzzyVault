# Reducciones de Tiempo

```python
dec = pd.read_csv('Huellas 106_6 y 106_7 dist 1800.csv')
dec.columns = ['Iteración real', 'Iteración permutada']

timesR = []
timesP = []
timesdiff = []
for i in range(len(dec)):
    timesR.append(((dec['Iteración real'][i]/1000)*23.5)/60)
    timesP.append(((dec['Iteración permutada'][i]/1000)*23.5)/60)
    timesdiff.append((timesP[i]-timesR[i]))
    
dec['Tiempo IR'] = timesR
dec['Tiempo Pr'] = timesP
dec['Dif Tiempo'] = timesdiff
    
# rt10md = rt10md.T.reset_index()
# rt10md.columns= ['Distancia', 'Puntos encontrados promedio', 'Puntos genuinos encontrados promedio', 'Número de emparejamientos que se podrían vulnerar', 'Iteraciones promedio necesarias para vulnerar']
# rt10md = rt10md.round({'Distancia':3, 'Puntos encontrados promedio':3, 'Puntos genuinos encontrados promedio':3, 'Número de emparejamientos que se podrían vulnerar':3, 'Iteraciones promedio necesarias para vulnerar':3})
dec
```
